# What is a Tensor

- A simple explanation would be that a tensor is a multi-dimensional array.




- A ***scalar*** is a single number.
- A ***vector*** is an array of numbers. 
- A ***matrix*** is a 2-dimensional array.
- A ***tensor*** is an n-dimensional array.

In TensorFlow, everything can be considered a tensor including a scalar. A scalar would be a tensor of dimension 0, a vector of dimension 1, and a matrix of dimension 2.

Now, this is useful because we are not limited to working with complex datasets in TensorFlow. TensorFlow can handle any type of data and feed it to machine learning models.

# How to Generate and Load Tensors

- Letâ€™s look at how to generate tensors.
- In most cases, you wonâ€™t be creating tensors from scratch. You will either load a dataset, convert other datasets like NumPy arrays to tensors, or generate tensors. First, letâ€™s look at how to generate tensors.

- Letâ€™s create a tensor with random values. There are two common ways you can do this: generate a normal distribution of data or a uniform distribution of data.


- The normal distribution is a bell-shaped curve that represents the distribution of data. Most of the data will be close to the average and fewer data will be away from the average. This means the probability of getting a value near the average is higher.




- The uniform distribution is a straight line that represents the distribution of data. All the values in a uniform distribution will have an equal probability of occurring within a given range.

- Before we generate random values, you must understand what a seed is. If we use a seed value, we can regenerate the same set of data multiple times. This will be useful when we want to test our machine-learning model against the same data after we tweak its performance

- A random seed is a starting point for generating random numbers using a random number generation algorithm. It's typically a number, but it could also be a set of numbers or even a random value itself. The important aspect of a random seed is that it initializes the random number generator in such a way that it produces a sequence of random numbers.


- Letâ€™s create two arrays of random tensors. We will first set a seed and generate the random values using that seed.

```python

seed = tf.random.Generator.from_seed(42)
seed
```

- tf.random.Generator.from_seed(42): This line creates a TensorFlow random number generator object (tf.random.Generator) with a specified seed value of 42.

- seed: This line likely intends to assign the newly created random number generator object to a variable named seed

- So, this code initializes a random number generator using TensorFlow with a seed value of 42. This means that every time you use this random number generator, it will produce the same sequence of random numbers, ensuring reproducibility.

* Now we will create a normal and uniform distribution with the shape of 3 by 2.

```python
normal_tensor = seed.normal(shape=(3,2))
print(normal_tensor)
uniform_tensor = seed.uniform(shape=(3,2))
print(uniform_tensor)
```

- This code snippet demonstrates how to generate random tensors using the TensorFlow library, with a specific random number generator initialized with a seed value of 42. Let's break it down step by step:

- normal_tensor = seed.normal(shape=(3,2)): This line generates a random tensor of shape (3, 2) filled with values from a normal distribution. The seed object, which is the random number generator initialized previously with a seed value of 42, is used to generate these random numbers. The normal method is used to specify that we want random numbers sampled from a normal (Gaussian) distribution.

- print(normal_tensor): This line prints the generated random tensor of shape (3, 2) filled with values from the normal distribution.

- uniform_tensor = seed.uniform(shape=(3,2)): This line generates another random tensor of shape (3, 2) filled with values from a uniform distribution. Again, the seed object is used to generate these random numbers, and the uniform method specifies that we want random numbers sampled from a uniform distribution.

- print(uniform_tensor): This line prints the generated random tensor of shape (3, 2) filled with values from the uniform distribution.

- In summary, this code snippet demonstrates how to use a specific random number generator initialized with a seed value to generate random tensors with TensorFlow, one sampled from a normal distribution and the other from a uniform distribution. The use of the seed ensures that the generated tensors will be the same each time the code is run, which is useful for reproducibility in scientific experiments or machine learning tasks.

- Next, we will create a tensor with zeros and ones. In TensorFlow, tensors filled with zeros or ones are often used as a starting point for creating other tensors. They can also be placeholders for inputs in a computational graph.

- To create a tensor of zeroes, use the tf.zeros function with a shape as the input argument. To create a tensor with ones, we use tf.ones with the shape as input argument.

```python
zeros = tf.zeros(shape=(3,2))
print(zeros)
ones = tf.ones(shape=(3,2))
print(ones)
```

- Now, letâ€™s look at converting NumPy arrays into tensors. If you donâ€™t know what NumPy is, it is a Python library for numerical computing. It helps us handle large datasets and perform a variety of computations on them.

- Letâ€™s import NumPy and create a NumPy array using NumPyâ€™s arrange function.

```python
import numpy as np
numpy_arr = np.arange(1,25,dtype=np.int32)
```

- This code snippet demonstrates the use of NumPy, a popular numerical computing library in Python, to create a NumPy array filled with integers. Here's the breakdown:

- import numpy as np: This line imports the NumPy library and assigns it the alias np. This alias is commonly used for brevity when working with NumPy functions and objects.

- numpy_arr = np.arange(1, 25, dtype=np.int32): This line creates a NumPy array called numpy_arr. The np.arange() function is used to generate a sequence of numbers. The parameters are:

- start: The starting value of the sequence (inclusive), which is 1 in this case.
- stop: The ending value of the sequence (exclusive), which is 25 in this case. So, the sequence will include numbers up to 24.
- dtype=np.int32: This specifies the data type of the array elements as 32-bit integers (np.int32). This means that each element in the array will be stored as a 32-bit integer.
- The np.arange() function generates an array containing the numbers from start to stop-1 with a step size of 1.

- So, after executing this code snippet, numpy_arr will be a NumPy array containing the integers from 1 to 24, with each element stored as a 32-bit integer.

- Now, we can create a tensor using the tf.constant function with the NumPy array as input. TensorFlow has built-in support to handle NumPy arrays, so it is just a matter of importing a NumPy array and setting a shape.

```python
print(numpy_arr)
numpy_tensor = tf.constant(numpy_arr,shape=[2,4,3])
print(numpy_tensor)
```

- numpy_tensor = tf.constant(numpy_arr, shape=[2, 4, 3]): This line creates a TensorFlow constant tensor named numpy_tensor. The tf.constant() function is used to create a constant tensor. It takes two main arguments:

- numpy_arr: The NumPy array that serves as the input data for creating the tensor.
- shape: The desired shape of the tensor. In this case, the shape is set to [2, 4, 3], which means the tensor will have dimensions 2x4x3.
- The input NumPy array will be reshaped into the specified shape to create the tensor.

- print(numpy_tensor): This line prints the TensorFlow tensor numpy_tensor that was created. The output shows the values of the tensor along with its shape (2, 4, 3) and data type int32.

- In summary, this code snippet demonstrates how to create a TensorFlow tensor from a NumPy array by using the tf.constant() function, providing the array as input data and specifying the desired shape of the tensor.

***You can see both the NumPy array as well as our tensor. The original NumPy array was 1x12 but our tensor is 2x4x3. This is called re-shaping a tensor which we will often do while training deep neural networks.***

# Basic Operations using Tensorflow

- We have learned how tensors are created in TensorFlow. Now letâ€™s look at some basic operations using tensors.

- We will start by getting some information on our tensors. Letâ€™s create a 4D tensor with 0 values with the shape 2x3x4x5.

```python
rank4_tensor = tf.zeros([2,3,4,5])
print(rank4_tensor)
```

- We have created our rank 4 tensor. Now let's get some information about the size, shape (number of values), and the dimension of the tensor.

- We will use tf.size function to get the size. The shape and ndim properties will give us the shape and dimensions of the tensor.

```python
print("Size",tf.size(rank4_tensor))
print("shape",rank4_tensor.shape)
print("Dimension",rank4_tensor.ndim)
```


- The shape of a tensor tells us the size of each dimension within the tensor. In the case of rank4_tensor, the shape is (2, 3, 4, 5). This means:

***The first dimension has a size of 2.***

***The second dimension has a size of 3.***


***The third dimension has a size of 4.***

***The fourth dimension has a size of 5***

```python
basic_tensor = tf.constant([[10,11],[12,13]])
print(basic_tensor)
```

Letâ€™s try some simple operations. We can add, subtract, multiply, and divide every value in a tensor using the basic operators.

```python
print(basic_tensor + 10)
print(basic_tensor - 10)
print(basic_tensor * 10)
print(basic_tensor / 10)
```

- - Now letâ€™s try matrix multiplication. I will create two simple tensors tensor_011 and tensor_012.

```python
tensor_011 = tf.constant([[2,2],[4,4]])
tensor_012 = tf.constant([[2,3],[4,5]])
```

- Keep in mind that in matrix multiplication, the inner dimensions should match. For example, a (3, 5) * (3, 5) multiplication wonâ€™t work but (3, 5) * (5, 3) will work.

- The final shape of the resulting matrix will be its outer dimension. so, a 3x5 tensor multiplied by a 5x3 tensor will give us a 5x5 tensor. We will use the tf.matmul function to perform matrix multiplication.

```python
print(tf.matmul(tensor_011,tensor_012))
```

For position (1,1) in the resulting matrix:

Row 1 of tensor_011 is [2, 2].
Column 1 of tensor_012 is [2, 4].
To find the element at position (1,1) in the resulting matrix, we perform the dot product of these two vectors:

(2 * 2) + (2 * 4) = 4 + 8 = 12


- Next, letâ€™s look at reshaping and transposing a matrix. As we saw before, we will often use reshaping to change our matrix structure while training neural networks.

- For example, an image pixel matrix of 28x28 will be converted into a 1-dimensional 784-pixel array for an image classification neural network.

- To reshape, we use the tf.reshape function. To transpose, we use the tf.transpose function. If you don't know what a transpose is, it's converting rows into columns and columns into rows.

```python
print(tf.reshape(tensor_011,[4,1]))
print(tf.transpose(tensor_011))
```

- Finally, letâ€™s look at some aggregate operations like min, max, standard deviation, square and square root.

- To find the minimum and maximum values, we use the tf.reduce_min and tf.reduce_max functions. And to find the sum of the array, we use the tf.reduce_sum function.

```python
tensor_013 = tf.constant([
    [1,2,3],
    [4,5,6],
    [7,8,9]
],dtype='float32')
print(tf.reduce_min(tensor_013))
print(tf.reduce_max(tensor_013))
print(tf.reduce_sum(tensor_013))
```

- Now for the standard deviation and variance, we use the tf.math.reduce_std function and tf.math.reduce_variance function.

```python
print(tf.math.reduce_std(tensor_013))
print(tf.math.reduce_variance(tensor_013))
```

- Letâ€™s find the square, square root, and log of each value in a tensor.

```python
print(tf.sqrt(tensor_013))
print(tf.square(tensor_013))
print(tf.math.log(tensor_013))
```
